Color Space Conversion

```python
import cv2
import os
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt


SRC_ROOT = Path("/content/Y2_S1_Batch02_Ku28/data/technic 3")
DST_ROOT = Path("/content/Y2_S1_Batch02_Ku28/data/technic 4")
SAMPLE_CLASS = "Apple___Apple_scab"
EXTS = (".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff")

DST_ROOT.mkdir(parents=True, exist_ok=True)

processed_count = 0
print(f"Starting: src={SRC_ROOT}\ndst={DST_ROOT}\n")

for src_dir, dirs, files in os.walk(SRC_ROOT):
    rel_dir = Path(src_dir).relative_to(SRC_ROOT)
    dst_dir = DST_ROOT / rel_dir
    dst_dir.mkdir(parents=True, exist_ok=True)

    for fname in files:
        if not fname.lower().endswith(EXTS):
            continue

        src_path = Path(src_dir) / fname
        dst_path = dst_dir / fname

        # read (OpenCV reads BGR)
        img_bgr = cv2.imread(str(src_path))
        if img_bgr is None:
            print(f" Skipping unreadable file: {src_path}")
            continue

        # convert to RGB for correct color-space ops / plotting
        img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)


        img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
        img_hsv  = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
        img_lab  = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2LAB)


        hsv_for_save = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB)
        save_bgr = cv2.cvtColor(hsv_for_save, cv2.COLOR_RGB2BGR)


        success = cv2.imwrite(str(dst_path), save_bgr)
        if not success:
            print(f" Failed to write: {dst_path}")

        processed_count += 1
        if processed_count % 200 == 0:
            print(f"Processed {processed_count} images...")

print(f"Done. Total images processed: {processed_count}")


# Visualize a sample
sample_dir = SRC_ROOT / SAMPLE_CLASS
if sample_dir.exists() and sample_dir.is_dir():
    files = [f for f in sample_dir.iterdir() if f.suffix.lower() in EXTS]
    if files:
        sample = files[0]
        img_bgr = cv2.imread(str(sample))
        img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
        gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
        hsv  = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
        lab  = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2LAB)

        # Convert HSV/LAB back to RGB for correct display with matplotlib
        hsv_disp = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
        lab_disp = cv2.cvtColor(lab, cv2.COLOR_LAB2RGB)

        fig, axs = plt.subplots(1, 4, figsize=(18, 5))
        axs[0].imshow(img_rgb); axs[0].set_title("Original (RGB)"); axs[0].axis("off")
        axs[1].imshow(gray, cmap="gray"); axs[1].set_title("Grayscale"); axs[1].axis("off")
        axs[2].imshow(hsv_disp); axs[2].set_title("HSV (visualized)"); axs[2].axis("off")
        axs[3].imshow(lab_disp); axs[3].set_title("LAB (visualized)"); axs[3].axis("off")
        plt.show()
    else:
        print(f"No images found in sample class folder: {sample_dir}")
else:
    print(f"Sample class folder missing: {sample_dir}  skip visualization.")
```

```python
import cv2
import os
from pathlib import Path
import numpy as np

# Paths
SRC_ROOT = Path("/content/Y2_S1_Batch02_Ku28/data/raw/Dataset")
DST_ROOT = Path("/content/Y2_S1_Batch02_Ku28/data/technic 5")
DST_ROOT.mkdir(parents=True, exist_ok=True)

# Supported extensions
EXTS = (".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff")

def reduce_color_depth(img, levels=8):
    factor = 256 // levels
    img_reduced = (img // factor) * factor
    return img_reduced.astype(np.uint8)

processed_count = 0

for src_dir, dirs, files in os.walk(SRC_ROOT):
    rel_dir = Path(src_dir).relative_to(SRC_ROOT)
    dst_dir = DST_ROOT / rel_dir
    dst_dir.mkdir(parents=True, exist_ok=True)

    for fname in files:
        if not fname.lower().endswith(EXTS):
            continue

        src_path = Path(src_dir) / fname
        dst_path = dst_dir / fname

        img_bgr = cv2.imread(str(src_path))
        if img_bgr is None:
            print(f" Skipping unreadable file: {src_path}")
            continue

        # Convert to RGB
        img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

        # Apply color depth reduction
        reduced_img = reduce_color_depth(img_rgb, levels=8)

        # Convert back to BGR for saving
        reduced_bgr = cv2.cvtColor(reduced_img, cv2.COLOR_RGB2BGR)

        # Save processed image
        cv2.imwrite(str(dst_path), reduced_bgr)

        processed_count += 1
        if processed_count % 200 == 0:
            print(f"Processed {processed_count} images...")

print(f" Done! Total processed: {processed_count}")
print(f"Saved to: {DST_ROOT}")
```

```python
import cv2
import matplotlib.pyplot as plt
from pathlib import Path

# CONFIG
IMG_PATH = Path("/content/Y2_S1_Batch02_Ku28/data/technic 5/Apple___Apple_scab/00075aa8-d81a-4184-8541-b692b78d398a___FREC_Scab 3335.JPG")

# Load image
img_bgr = cv2.imread(str(IMG_PATH))
if img_bgr is None:
    raise FileNotFoundError(f" Could not read image: {IMG_PATH}")

# Convert to RGB
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Plot histogram
colors = ('r','g','b')
plt.figure(figsize=(10,5))
for i, col in enumerate(colors):
    hist = cv2.calcHist([img_rgb], [i], None, [256], [0,256])
    plt.plot(hist, color=col)
    plt.xlim([0,256])

plt.title("RGB Histogram")
plt.xlabel("Pixel intensity")
plt.ylabel("Frequency")
plt.show()
```
